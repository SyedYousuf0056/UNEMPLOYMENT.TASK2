```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Load the uploaded CSV file
df = pd.read_csv("Unemployment_in_India.csv")

# Show the first 5 rows
df.head()
```

```python
# Show basic info about the dataset
df.info()

# Check for missing/null values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Set the figure size
plt.figure(figsize=(12,6))

# Create bar plot
sns.barplot(x='Region', y='Estimated Unemployment Rate (%)', data=df)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Title and labels
plt.title("Unemployment Rate by Region (May 2020)")
plt.xlabel("Region")
plt.ylabel("Unemployment Rate (%)")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate correlation matrix
correlation = df.corr(numeric_only=True)

# Set up the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap="YlGnBu", linewidths=0.5)

# Add title
plt.title("Correlation Heatmap of Unemployment Data")

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Sort the dataframe by Unemployment Rate in descending order
top_unemployment = df.sort_values(by="Estimated Unemployment Rate (%)", ascending=False)

# Display the top 5 regions with the highest unemployment
top_unemployment[["Region", "Estimated Unemployment Rate (%)"]].head()
```

```python
# Save the processed dataframe to a new CSV file
df.to_csv("Cleaned_Unemployment_Data.csv", index=False)

print("Cleaned data saved as 'Cleaned_Unemployment_Data.csv'")
```
